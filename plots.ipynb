```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
sns.set_style('dark')
main = '#25a6d9'
housing = pd.read_csv("D:/Lern ht code/begin-python/first_analysis_Pyth/Housing.csv")
print(housing)
```

```python
#### 1.1 Bar plot
plt.figure(figsize=(6, 3)) # было (10, 6)
sns.barplot(data=housing, x='bedrooms', y='area', color=main) # plt.bar(housing['bedrooms'], housing['price'])
plt.xlabel("area")
plt.ylabel("bedrooms")

plt.show()
```

```python
#### 1.2 Hist plot
plt.figure(figsize=(6, 3)) # было (10, 6)
sns.histplot(data=housing, x='price', log_scale=True, bins=20, color=main) #plt.hist(housing['price'], bins=20)
plt.xlabel("price")
plt.ylabel("amount")
plt.show()
```

```python
#### 1.3 Boxplot

plt.figure(figsize=(6, 3)) # было (10, 6)
plt.boxplot(housing['price'])
plt.show()

# результат показывает что наблюдения от 25% до 75% находятся в пределе от около 0.3 до 0.55. медианная цена находится выше 0.4
```

```python
#### 1.4 Scatter Plot

plt.figure(figsize=(6, 3)) # было (10, 6)
sns.scatterplot(data=housing, x='area',y='price', color=main) #plt.scatter(housing['area'], housing['price'])
plt.show()

corr_price_area = np.corrcoef(housing['price'], housing['area'])
print(corr_price_area)
# График - показывает положительную корреляцию, но она не сильная
# Функция np.corrcoef() это подтверждает, корреляция равняется 0.536
```

```python
#### 1.5. Line Plot

sns.lineplot(x='area', y='price', data=housing)
plt.title('Цена от площади')
plt.xlabel('Площадь')
plt.ylabel('Цена')
plt.show()
# по графику видно, что с увеличением площади цена возрастает, но не сильно
```

```python
#### 1.7 Heatmap
numeric_cols = housing.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(6, 4)) 
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.xticks(rotation=0, fontsize=9)
plt.yticks(fontsize=9)
plt.show()
```

```python
plt.figure(figsize=(6, 3)) # было (10, 6)
sns.pairplot(housing)
plt.title('Pair Plot Example')
plt.show()
# не сильно понимаю часть этих графиков, но они есть
```
